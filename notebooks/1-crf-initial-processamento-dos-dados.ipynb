A <font color=82D712>**Alura Voz**</font>, atua na área de telecomunicações através de serviços de telefônia e como Internet Service Provider (ISP), com diversas tecnologias de internet (DNS e Fibra Otica). No mercado desde 1999, a empresa atraiu diversos clientes e parceiros na área corporativa e de telefonia celular, e se firmou como forte nome nacional em soluções que integravam internet e internet móvel, apostando em sua capacidade de entrega de soluções altamente inovadora. Porém, recentemente vem apresentando uma <font color=82D712>*alta taxa de churn*</font> entre seus clientes, que tem prejudicado seu crescimento. <font color=82D712><strong>*</strong></font>
<br>
A **taxa de churn**, também conhecida como churn de clientes ou taxa de churn, é a taxa na qual os clientes param de fazer negócios com uma entidade. É mais frequentemente expresso como a porcentagem de assinantes do serviço que *pararam de assinar* durante um período de tempo específico. Para uma empresa expandir sua base de clientes, **sua taxa de crescimento** (medida pelo número de novos clientes) deve exceder **sua taxa de churn**. Desse modo, cabe ao cientista de dados responsavel, analizar as informações coletadas no banco de dados da empresa e destacar as informações necessarias, de modo, a auxiliar a tomada de deciçoes da empresa.<br>

# <font color=2C3E50>**02. Impotação e Analise Inicial dos dados**

```python
import pandas as pd
import pandas_profiling
import numpy as np
```

<font color=82D712> **Analizando os Dados**

```python
API_Data = pd.read_json('/content/Telco-Customer-Churn.json')
API_Data.head(2)
```

As colunas do dataframe possuem dicionários como valores, logo, será utilizado o comando seguinte comando para este fim: <br>
> `<font color=82D712>pd.json_normalize</font>`

Esse código trasformará esses dicionários em novos dataframes, onde será então feito uma nova analize dos dados.

## <font color=2C3E50>**02.1. Analizando os Dados: <font color=82D712>Clientes**

```python
clientes = pd.json_normalize(data=API_Data['customer'])
clientes.head()
```

```python
mapa_clientes = {'gender':'Sexo','SeniorCitizen':'Idoso',
                 'Partner':'Parceiros','Dependents':'Dependentes',
                 'tenure':'Tenure'}
                 
clientes = clientes.rename(columns = mapa_clientes)
clientes.head()
```

<font color=82D712> **Analizando as variavéis únicas desse dataframe ...**

```python
for col in clientes:
   print("="*40)
   print('Valores Únicos da Coluna:', col)
   print('')
   print(clientes[col].unique())
   print('')
```



> Os <font color=82D712>**dados Binários**</font> (Yes/No)(Female/Male) podem gerar problemas em futuras analises, logo, trasformaremos em 0 e 1.

```python
correcao_clientes = ['Sexo', 'Parceiros', 'Dependentes',]

for column in correcao_clientes:
  clientes[column] = clientes[column].replace({"Male":0, "Female": 1, 'No':0, 'Yes':1})
```

```python
clientes.head()
```

## <font color=2C3E50>**02.2. Analizando os Dados: <font color=82D712>Telefone**

```python
telefone = pd.json_normalize(data=API_Data['phone'])
telefone.head()
```

```python
mapa_telefone = {'PhoneService':'Telefonia',	'MultipleLines':'Multiplas_Linhas'}
                 
telefone = telefone.rename(columns = mapa_telefone)
telefone.head()
```

<font color=82D712> **Analizando as variavéis únicas desse dataframe ...**

```python
for col in telefone:
   print("="*50)
   print('Valores Únicos da Coluna:', col)
   print('')
   print(telefone[col].unique())
   print('')
```

> Os <font color=82D712>**dados Binários**</font> (Yes/No) podem gerar problemas em futuras analises, logo, seram trasformadas em 0 e 1. Além disso, pode-se afirmar que os dados **'No phone service'** e **'No'** possuem mesma significado (no contexto dessa analise), então seram substituido tambem por 0.

```python
correcao_telefone = ['Telefonia',	'Multiplas_Linhas']

for column in correcao_telefone:
  telefone[column] = telefone[column].replace({'No':0, 'No phone service':0, 'Yes':1})
```

```python
telefone.head()
```

## <font color=2C3E50>**02.3. Analizando os Dados: <font color=82D712>Internet**

```python
internet = pd.json_normalize(data=API_Data['internet'])
internet.head()
```

```python
mapa_internet = {'InternetService':'Internet',	'OnlineSecurity':'Segurança_On-line',	
                 'OnlineBackup': 'Backup_On-line',	'DeviceProtection':'Proteção_Dispositivos',	
                 'TechSupport':'Suporte_Técnico',	'StreamingTV':'TV_Streaming',	'StreamingMovies':'Streaming_Filmes'}
                 
internet = internet.rename(columns = mapa_internet)
internet.head()
```

<font color=82D712> **Analizando as variavéis únicas desse dataframe ...**

```python
for col in internet:
   print("="*50)
   print('Valores Únicos da Coluna:', col)
   print('')
   print(internet[col].unique())
   print('')
```

> Os <font color=82D712>**dados Binários**</font> (Yes/No) podem gerar problemas em futuras analises, logo, seram trasformadas em 0 e 1. Além disso, pode-se afirmar que os dados 'No internet service' e 'No' possuem mesma significado (no contexto dessa analise), então seram substituido tambem por 0 (no)

```python
correcao_internet = ['Segurança_On-line',	'Backup_On-line',	'Proteção_Dispositivos',	
                     'Suporte_Técnico',	'TV_Streaming',	'Streaming_Filmes']

for column in correcao_internet:
  internet[column] = internet[column].replace({'No':0, 'No internet service':0, 'Yes':1})
```

```python
internet.head()
```

> Os dados que possuem mais de duas variáveis, também podem gerar problemas em futuras análises, mas, não podem ser trasformadas em 0 e 1 . Desse modo, será usado o metodo de <font color=82D712>**One-Hot-Enconding**</font>, que fará cada variável uma nova coluna do tipo booleano(0 ou 1)

```python
internet = pd.get_dummies(data=internet, columns=['Internet'])
internet.head()
```

```python
correcao_internet_2 = {'Internet_No':'Sem_internet'}

internet = internet.rename(columns = correcao_internet_2)
internet.head()
```

## <font color=2C3E50>**02.4. Analizando os Dados: <font color=82D712>Conta**

```python
conta = pd.json_normalize(data=API_Data['account'])
conta.head()
```

```python
mapa_conta = {'Contract':'Contrato',	'PaperlessBilling':'Pagamento_Eletrônico', 'PaymentMethod':'Método_Pagamento',
              'Charges.Monthly':'Cobrança_Mensal', 'Charges.Total':'Cobrança_Total'}
                 
conta = conta.rename(columns = mapa_conta)
conta.head()
```

<font color=82D712> **Analizando as variavéis únicas desse dataframe ...**

```python
for col in conta:
   print("="*60)
   print('Valores Únicos da Coluna:', col)
   print('')
   print(conta[col].unique())
   print('')
```

> Os <font color=82D712>**dados Binários**</font> (Yes/No) podem gerar problemas em futuras analises, logo, seram trasformadas em 0 e 1.
```python
conta.head()
```

## <font color=2C3E50>**03.2. Corrigindo inconsistência no : <font color=82D712>Tenure**

Quando se analizou as var com o pandas_profile foi identificado que o valor minimo da variável <font color=82D712>**tenure**</font> é 0. Porém, para este tipo de dado (tempo de contrato) é estranho um cliente já que representa:

> ___"O número de meses que um Devedor teve uma conta com base na data de estabelecimento de conta ativa mais antiga para tal Devedor"___ <sup>[[1]](https://www.lawinsider.com/dictionary/customer-tenure).

Então, será necessário entender melhor esse parâmetro para determinar o que este dado representa e se é uma inconsistência. <br>
Também é identificado no relatório que o **<font color=82D712>tenure</font>** é altamente relacionado com o tipo de contrato, em especial, o ***mensal*** e o ***bianual***, então vamos analizar os dados 0 considerando esses cenários. 

```python
tenure = pd.DataFrame(data=df, columns=['Tenure', 'Cobrança_Mensal', 'Cobrança_Total', 
                                        'Contrato_Mensal', 'Contrato_Anual', 'Contrato_Bianual'])
```

```python
tenure.query('Tenure == 0')
```

```python
tenure.query('Contrato_Bianual == 1').shape[0]
```

```python
tenure.query('Contrato_Bianual == 1').query('Tenure == 48').head()
```

---
<font color=82D712 size="5">**Conclusões**</font>

* Pode-se observar que todos que possuem tenure igual a 0 também possuem contrato do tipo Bianual/Anual, mas, que nem toda pessoa com esse tipo de contrato possui esse tenure, logo, não se pode afirmar que o tenure 0 é relacionado ao tipo de contrato. <br>
* Além disso, as linhas de tenure 0, tem suas cobranças totais vazias, mas, possum valores pagos nas cobranças mensais.

Logo, esses dados podem significar três coisas:
1. 0 representa ***o primeiro mês do contrato*** e a cobrança total tem valor 0, pois, ainda não foi cobrado a mensalidade.
2. A empresa pode ofertar um serviço de ***avaliação gratuita***, por isso no primeiro mês a cobrança é 0. 
3. Houve algum ***erro*** durante o armazenamento desses dados

De qualquer forma, não afetam de forma drastica a variável target e outras informações presentes nessas linhas podem ser uteis futuramente. Logo, seram mantidas dessa forma.
